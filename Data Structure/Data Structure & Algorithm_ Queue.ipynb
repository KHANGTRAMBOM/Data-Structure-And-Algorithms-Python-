## **2. C·∫•u tr√∫c & C√°c c√°ch c√†i ƒë·∫∑t**

`Queue` c√≥ th·ªÉ ƒë∆∞·ª£c c√†i ƒë·∫∑t b·∫±ng nhi·ªÅu c√°ch, ph·ªï bi·∫øn nh·∫•t l√† s·ª≠ d·ª•ng **m·∫£ng (array)** v√† **danh s√°ch li√™n k·∫øt (linked list)**. Trong ph·∫ßn n√†y, ch√∫ng ta s·∫Ω t·∫≠p trung v√†o vi·ªác c√†i ƒë·∫∑t b·∫±ng danh s√°ch li√™n k·∫øt.

*   **S·ª≠ d·ª•ng Danh s√°ch li√™n k·∫øt (Linked List):**
    *   M·ªói ph·∫ßn t·ª≠ trong `Queue` l√† m·ªôt `node` c·ªßa danh s√°ch li√™n k·∫øt.
    *   ƒê·ªÉ qu·∫£n l√Ω, ta d√πng hai con tr·ªè: `front` tr·ªè ƒë·∫øn ƒë·∫ßu v√† `rear` tr·ªè ƒë·∫øn cu·ªëi h√†ng ƒë·ª£i.
    *   **Enqueue (Th√™m v√†o):** Th√™m m·ªôt `node` m·ªõi v√†o `cu·ªëi` danh s√°ch (c·∫≠p nh·∫≠t con tr·ªè `rear`).
    *   **Dequeue (L·∫•y ra):** X√≥a `node` ·ªü `ƒë·∫ßu` danh s√°ch (c·∫≠p nh·∫≠t con tr·ªè `front`).
    *   **Peek (Xem):** Tr·∫£ v·ªÅ gi√° tr·ªã c·ªßa `node` ·ªü `ƒë·∫ßu` (`front`) m√† kh√¥ng x√≥a n√≥.

# **CODE**

Trong ph·∫ßn n√†y, t√¥i s·∫Ω vi·∫øt l·∫°i c·∫•u tr√∫c d·ªØ li·ªáu `Queue` t·ª´ ƒë·∫ßu (from scratch) b·∫±ng ng√¥n ng·ªØ Python, m√† **kh√¥ng s·ª≠ d·ª•ng b·∫•t k·ª≥ th∆∞ vi·ªán h·ªó tr·ª£ n√†o**.

Notebook n√†y ch·ªâ mang t√≠nh ch·∫•t tham kh·∫£o v√† ph·ª•c v·ª• m·ª•c ƒë√≠ch h·ªçc t·∫≠p. N·∫øu c√≥ sai s√≥t g√¨, r·∫•t mong nh·∫≠n ƒë∆∞·ª£c s·ª± g√≥p √Ω v√† ƒë√°nh gi√° t·ª´ c√°c b·∫°n üòåüòåüòå


```python
# T·∫°o node c∆° b·∫£n
class Node:
  def __init__(self, val: int, next = None):
    self.val = val
    self.next = next
```

```python
# T·∫°o Queue
class Queue:
  # G·ªìm 2 th√†nh ph·∫ßn ch√≠nh front/head v√† rear/back th√†nh ph·∫ßn ph·ª• size ƒë·ªÉ l·∫•y k√≠ch th∆∞·ªõc Queue v·ªõi O(1)
  def __init__(self):
    self.__front = None
    self.__rear = None
    self.size = 0

  # Ki·ªÖm tra r·ªóng
  def isEmpty(self) -> bool:
    return not self.__front

  # Th√™m m·ªôt ph·∫ßn t·ª´ v√†o cu·ªëi Queue
  def Enqueue(self, val:int) -> None:
    new_node = Node(val)

    # N·∫øu ƒëang r·ªóng th√¨ g√°n l·∫°i cho front v√† rear
    if self.isEmpty():
      self.__front = new_node
      self.__rear = self.__front
    # N·∫øu kh√¥ng th√¨ ƒë∆°n gian rear.next tr·ªè v√†o ph·∫ßn t·ª´ m·ªõi v√† g√°n l·∫°i rear
    else:
      self.__rear.next = new_node
      self.__rear = self.__rear.next
    self.size += 1

  def Dequeue(self) -> int:
    # N·∫øu Queue kh√¥ng r·ªóng th√¨ l·∫•y gi√° tr·ªã ƒë·∫ßu r·ªìi x√≥a n√≥
    if not self.isEmpty():
        result = self.__front.val
        self.__front = self.__front.next
        self.size -= 1
        if self.size == 0:
          self.__rear = None

        return result
    else:
      raise IndexError("Queue hi·ªán t·∫°i ƒëang r·ªóng")

  # Xem gi√° tr·ªã ƒë√†u m√† kh√¥ng c·∫ßn x√≥a
  def Peek(self) -> int:
    if not self.isEmpty():
      return self.__front.val

  # Hi·ªÉn th·ªã c√°c ph·∫ßn t·ª≠ c√≥ trong Queue
  def toString(self) -> None:
    tmp = self.__front
    while tmp:
      print(f"{tmp.val} ",end ="")
      tmp = tmp.next
    print()
```

```python
### Test case ch·ª©c nƒÉng th√™m
myQueue = Queue()
myQueue.Enqueue(1)
myQueue.Enqueue(2)
myQueue.Enqueue(3)
myQueue.Enqueue(4)
myQueue.Enqueue(5)
myQueue.toString()
```

```python
### Test case ch·ª©c nƒÉng x√≥a v√† xem
print("Tr∆∞·ªõc khi x√≥a: ", end = "")
myQueue.toString()
print("Sau khi x√≥a: ", end = "")
myQueue.Dequeue()
myQueue.toString()
print("T·ªïng s·ªë l∆∞·ª£ng ph·∫ßn t·ª´ c√≤n l·∫°i l√†: ", myQueue.size)
print("Ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n hi·ªán t·∫°i l√†: ", myQueue.Peek())
```

## [LEETCODE 933. Number of Recent Calls](https://leetcode.com/problems/number-of-recent-calls/description) **`EASY`**

You have a RecentCounter class which counts the number of recent requests within a certain time frame.

Implement the RecentCounter class:

- RecentCounter() Initializes the counter with zero recent requests.
- int ping(int t) Adds a new request at time t, where t represents some time in milliseconds, and returns the number of requests that has happened in the past 3000 milliseconds (including the new request). Specifically, return the number of requests that have happened in the inclusive range [t - 3000, t].

It is guaranteed that every call to ping uses a strictly larger value of t than the previous call.

```python
### B√†i to√°n n√†y y√™u c·∫ßu ·ªü l·∫ßn ping g·∫ßn nh·∫•t v√† x√©t trong 3 gi√¢y tr∆∞·ªõc ƒë√≥ c√≥ bao nhi√™u ping h·ª£p l·ªá
### gi√° s·ª≠ ta ping(1) ping(100) ping(3001) ping(3002). ·ªû l·∫ßn ping ping(3001) ta c√≥ th·ªÉ th·∫•y c√≥ 3 ping h·ª£p l·ªá
### l√† ping(1) ping(100) ping(3001) c√≤n ·ªü ping(3002) ta th·∫•y ping(1) kh√¥ng h·ª£p l·ªá n√™n khi ƒë√≥ 3 ping h·ª£p l·ªá
### s·∫Ω l√† ping(100) ping(3001) ping(3002).
### Ph∆∞∆°ng ph√°p ti·∫øp c·∫≠n trong b√†i n√†y ch√≠nh l√† so s√°nh th·ªùi gian ping ƒë·∫ßu ti√™n v·ªõi l·∫ßn ping g·∫ßn nh·∫•t, t∆∞∆°ng
### ·ª©ng l√† so s√°nh gi√° tr·ªã front v·ªõi rear trong queue, v√† queue ch√≠nh l√† ch√¨a kh√≥a c·ªßa b√†i n√†y

class RecentCounter:

    def __init__(self):
        self.count = 0
        self.queue = []

    def ping(self, t: int) -> int:
        # Th·ª±c hi·ªán l·∫≠p cho ƒë·∫øn khi h·ª£p l·ªá
        while True:
            # N·∫øu ban ƒë·∫ßu l√† r·ªóng th√¨ th√™m v√†o kh√¥ng x√©t th√™m
            if not self.queue:
                self.queue.append(t)
                break

            # Ti·∫øn h√†nh l·∫•y gi√° tr·ªã ƒë·∫ßu ti√™n (front) ƒë·ªÉ so s√°nh
            top = self.queue[0]

            # H·ª£p l·ªá th√¨ th√™m v√†o
            if t - top <= 3000:
                self.queue.append(t)
                break
            # Kh√¥ng h·ª£p l·ªá th√¨ x√≥a n√≥ kh·ªèi queue v√† ti·∫øn h√†nh so s√°nh ti·∫øp
            else:
                self.queue.pop(0)
                self.count -= 1

        self.count += 1
        return self.count
```

## [LEETCODE 622. Design Circular Queue](https://leetcode.com/problems/design-circular-queue/description) **`MEDIUM`**

Design your implementation of the circular queue. The circular queue is a linear data structure in which the operations are performed based on FIFO (First In First Out) principle, and the last position is connected back to the first position to make a circle. It is also called "Ring Buffer".

One of the benefits of the circular queue is that we can make use of the spaces in front of the queue. In a normal queue, once the queue becomes full, we cannot insert the next element even if there is a space in front of the queue. But using the circular queue, we can use the space to store new values.

Implement the MyCircularQueue class:

- MyCircularQueue(k) Initializes the object with the size of the queue to be k.
- int Front() Gets the front item from the queue. If the queue is empty, return -1.
- int Rear() Gets the last item from the queue. If the queue is empty, return -1.
- boolean enQueue(int value) Inserts an element into the circular queue. Return true if the operation is successful.
- boolean deQueue() Deletes an element from the circular queue. Return true if the operation is successful.
- boolean isEmpty() Checks whether the circular queue is empty or not.
- boolean isFull() Checks whether the circular queue is full or not.

You must solve the problem without using the built-in queue data structure in your programming language.




```python
### B√†i n√†y y√™u c·∫ßu ch√∫ng ta x√¢y d·ª±ng queue d·ª±a v√†o circular array (m·∫£ng v√≤ng). L√Ω do
### b·ªüi v√¨ queue ta lu√¥n ƒë∆∞a gi√° tr·ªã v√†o cu·ªëi cho n√™n queue ƒëang ƒë·∫ßy v√† ta dequeue th√¨
### c√°c gi√° tr·ªã ƒë·∫ßu s·∫Ω r·ªóng, m·∫∑c d√π l√† r·ªóng nh∆∞ng kh√¥ng th·ªÉ enqueue th√™m ƒë∆∞·ª£c. Cho n√™n
### ch√∫ng ta d√πng m·∫£ng v√≤ng ƒë·ªÉ c√≥ th·ªÉ t·∫≠n d·ª•ng b·ªô nh·ªõ c·ªßa ch√∫ng ta

### H∆∞·ªõng ti·∫øp c·∫≠n c·ªßa b√†i to√°n n√†y ch√≠nh l√† ph√©p tinh chia d∆∞. Ph√©p t√≠nh chia d∆∞ l√†
### ph√©p t√≠nh mang t√≠nh v√≤ng l·∫≠p b·ªüi v√¨ gi√° tr·ªã c·ªßa n√≥ kh√¥ng bao gi·ªù l∆°n h∆°n s·ªë chia
### v√≠ d·ª• 4 % 5 d∆∞ 4 v√† 5 % 5 d∆∞ 0. N√≥ v√¥ t√¨nh kh·ªõp v·ªõi ch·ªâ m·ª•c trong m·∫£ng ƒëi t·ª´ 0
### ƒë·∫øn n - 1 gi·ªëng nh∆∞ k·∫øt qu·∫£ c·ªßa ph√©p chia d∆∞.

class MyCircularQueue:
    def __init__(self, k: int):
        self.list  = [0] * k
        self.capacity = k
        self.front = 0
        self.rear =  0
        self.count = 0

    def enQueue(self, value: int) -> bool:
        # N·∫øu ƒë√£ ƒë·∫ßy th√¨ kh√¥ng th·ªÉ th√™m
        if self.isFull():
            return False

        # N·∫øu ƒëang r·ªóng th√¨ front v√† queue c√πng tr·ªè v·ªÅ m·ªôt ch·ªï
        if self.isEmpty():
            self.list[self.front] = value
            self.rear = self.front
        # Kh√¥ng th√¨ th√™m v√†o cu·ªëi ƒë·ªìng th·ªùi c·∫≠p nh·∫≠t l·∫°i rear
        else:
            self.rear = (self.rear + 1) % self.capacity
            self.list[self.rear] = value

        self.count += 1

        return True

    def deQueue(self) -> bool:
        # Ch·ªâ d∆°n gi·∫£n l√† ƒë∆∞a front th√†nh ph·∫ßn t·ª≠ ti·∫øp theo
        if self.isEmpty():
            return False
        self.front = (self.front + 1) % self.capacity
        self.count -= 1
        return True

    # C·∫£ Front v√† Rear s·∫Ω tr·∫£ v·ªÅ gi√° tr·ªã ·ªü ƒë·∫ßu v√† cu·ªëi ·ªü hi·ªán t·∫°i n·∫øu r·ªóng th√¨
    # tr·∫£ l·∫°i -1
    def Front(self) -> int:
        return self.list[self.front] if not self.isEmpty() else -1

    def Rear(self) -> int:
        return self.list[self.rear]  if not self.isEmpty() else -1

    def isEmpty(self) -> bool:
        return self.count == 0

    def isFull(self) -> bool:
        return self.count == self.capacity
```
